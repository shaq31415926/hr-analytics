```python
# will drop the first one and rename - not quite sure what the first gender column is

books.drop(['gender'], inplace = True, axis = 1)

books = books.rename(columns = {"gender_1": "gender"})
```

```python
books.gender.value_counts()
```

```python
# identify missing data

books.isnull().sum()

# which variables are completely missing? The output shows number of missing values. 0 means no missing values.
```

```python
# cases where job_title_other is not null
# will drop this column

books[~books.job_title_other.isnull()]
```

```python
# remove job title other as contains missing data and does not add any further information

books.drop(["job_title_other"], axis = 1, inplace = True) 
```
